## Function

A function for plotting that we are going to use later on.

````python
def plot_clusters(data, centroids):
    """
    Shows a scatter plot with the data points clustered according to the centroids.
    """
    # Assigning the data points to clusters/centroids.
    clusters = [[] for _ in range(centroids.shape[0])]
    for i in range(data.shape[0]):
        distances = np.linalg.norm(data[i] - centroids, axis=1)
        clusters[np.argmin(distances)].append(data[i])

    # Plotting clusters and centroids.
    fig, ax = plt.subplots()
    for c in range(centroids.shape[0]):
        if len(clusters[c]) > 0:
            cluster = np.array(clusters[c])
            ax.scatter(cluster[:, 0], cluster[:, 1], s=7)
    ax.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, c='red')
````

## Implementation

A human can with an ease find five distinct clusters just by watching the plot. A computer, however, needs to be told how to find the clusters.

**Exercise:**

Implement the k-means clustering algorithm.

* Use the Euclidean (L<sub>2</sub>) distance.
* It is sufficient to use the basic Python constructs in your implementation, even though we heavily rely on Numpy throughout this assignment.

````python
def kmeans(data, centroids):
    """Function implementing the k-means clustering.

    Args:
        data ([[float]]): List of (x,y) coordinates of data points
        centroids ([[float]]): List of (x,y) coordinates of centroids

    Returns:
        [[float]]: List of (x,y) coordinates of centroids
    """

    # We'll need to initialize euclidean distance for all data points to the different centroids
    # Thinking a list of length same as data points, with distance to each centroid in a list
    euclidean_distance_to_centroids = np.zeros((data.shape[0], centroids.shape[0]))
    prev_centroids = centroids.copy()

    # Have to iteratively run kmeans clustering until the clusters don't change
    while True:
        new_centroids = prev_centroids.copy()

        # Tried using the l2 norm as this should give the euclidean distance, but did not get the correct distances, so have to have done something wrong
        # Can compute the square root of the sum of the squares of the differences between the values instead
        """ for i, current_datapoint in enumerate(data):
            for c, current_centroid in enumerate(new_centroids):
                euclidean_distance_to_centroids[i, c] = np.sum(np.sqrt(np.square(current_datapoint - current_centroid))) 
        """
        # Simplify the code over columns with axis=1
        for i, centroid in enumerate(new_centroids):
            euclidean_distance_to_centroids[:, i] = np.sum(np.sqrt(np.square(data - centroid)), axis=1)

        # Now we want to find which of the centroid each data point is closest to
        closest_cluster_per_datapoint = np.zeros_like(euclidean_distance_to_centroids) # Array of 0s and 1s indicating closest cluster based on index
        for i, distances in enumerate(euclidean_distance_to_centroids):
            index_of_min_value = np.argmin(distances)
            closest_cluster_per_datapoint[i][index_of_min_value] = 1

        # Last step is to take mean of the closest data points based on each centroid
        # Therefore we have to filter data on the closest_cluster_per_datapoint == 1
        for i in range(len(new_centroids)):
            new_centroids[i] = np.mean(data[closest_cluster_per_datapoint[:,i] == 1], axis=0)

        if(np.array_equal(prev_centroids, new_centroids)):
            centroids = new_centroids
            break
        
        prev_centroids = new_centroids.copy()
    
    return centroids
````

## Evaluation of Clustering

Silhouette Coefficient is an example of a measure for validation of the cluster quality.

**Exercise:**

Implement a function calculating the mean Silhouette Coefficient of all samples.

* Use the Euclidean (L<sub>2</sub>) distance.
* It is sufficient to use the basic Python constructs in your implementation, even though we heavily rely on Numpy throughout this assignment.

````python
def silhouette_score(data, centroids):
    """Function implementing the k-means clustering with the silhouette score

    Args:
        data ([[float]]): List of (x,y) coordinates of data points
        centroids ([[float]]): List of (x,y) coordinates of centroids

    Returns:
        float: mean Silhouette Coefficient of all samples
    """
    ### START CODE HERE ###
    silhoutte_scores = np.zeros(len(data))
    
    # Copy code from kmeans method, could have made method here
    euclidean_distance_to_centroids = np.zeros((data.shape[0], centroids.shape[0]))
    # Simplify the code over columns with axis=1
    for i, centroid in enumerate(centroids):
        euclidean_distance_to_centroids[:, i] = np.sum(np.sqrt(np.square(data - centroid)), axis=1)

    # Now we want to find which of the centroid each data point is closest to
    closest_cluster_per_datapoint = np.zeros_like(euclidean_distance_to_centroids) # Array of 0s and 1s indicating closest cluster based on index
    for i, distances in enumerate(euclidean_distance_to_centroids):
        index_of_min_value = np.argmin(distances)
        closest_cluster_per_datapoint[i][index_of_min_value] = 1
    
    for c, centroid in enumerate(centroids):
        current_cluster = data[closest_cluster_per_datapoint[:,c] == 1] 
        
        # First calculate a
        a = np.zeros(len(current_cluster))
        for point in range(len(current_cluster)): # for every point in the cluster
            a[point] = np.mean(np.sqrt(np.sum(np.square(current_cluster[point] - current_cluster), axis=1)))
        
        # Then calculate b
        b = np.zeros(len(current_cluster))
        # For each point in the current cluster
        for point in range(len(current_cluster)):
            lowest_dist_outside_cluster = np.Inf
            # For every cluster outside the current cluster
            for c_not_centroid, other_centroids in enumerate(centroids):
                if not np.array_equal(other_centroids, centroid):
                    cluster = data[closest_cluster_per_datapoint[:, c_not_centroid] == 1]
                    cluster_dist = np.mean(np.sqrt(np.sum(np.square(current_cluster[point] - cluster), axis=1)))
                    if cluster_dist < lowest_dist_outside_cluster:
                        b[point] = cluster_dist
                        lowest_dist_outside_cluster = cluster_dist
        silhoutte_scores[closest_cluster_per_datapoint[:, c] == 1] = (b - a)/np.maximum(a, b)
    score = np.mean(silhoutte_scores)

    ### END CODE HERE ### 
    return score
````

## Comments

Our k-means clustering implementation can be characterized as a naive. This is for following reasons:

* We are evaluating only one `k` value instead of trying multiple.
* We are initializing the centroids randomly instead of using some heuristic.
* We are initializing and evaluating only one set of centroids instead of initializing multiple sets and analyzing their SSE (Sum of Squared Errors) or Silhouette Coefficient.
